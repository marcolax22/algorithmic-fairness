```python
import numpy as np
import matplotlib.pyplot as plt

# Define the selection probabilities for two groups (e.g., high and low qualification)
p_high_qual = 0.8  # 80% chance of selection for high qualification
p_low_qual = 0.4   # 40% chance of selection for low qualification

# Number of candidates to simulate in each run
n_candidates = 1000

# Number of Monte Carlo simulations (iterations)
n_simulations = 10000

# Create arrays to store hiring outcomes for each simulation
high_qual_outcomes = np.zeros(n_simulations)
low_qual_outcomes = np.zeros(n_simulations)

# Run the Monte Carlo simulation
for i in range(n_simulations):
    # Simulate the hiring process for each group using a Bernoulli distribution
    high_qual_selected = np.random.binomial(1, p_high_qual, n_candidates)
    low_qual_selected = np.random.binomial(1, p_low_qual, n_candidates)
    
    # Calculate the final hiring rates for each group
    high_qual_outcomes[i] = np.mean(high_qual_selected)
    low_qual_outcomes[i] = np.mean(low_qual_selected)

# Plot the results: Distribution of hiring rates for both groups
plt.figure(figsize=(10, 6))

# Plot histogram for high qualification group
plt.hist(high_qual_outcomes, bins=50, alpha=0.5, label='High Qualification', color='blue')

# Plot histogram for low qualification group
plt.hist(low_qual_outcomes, bins=50, alpha=0.5, label='Low Qualification', color='red')

# Add labels and title
plt.xlabel('Hiring Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Hiring Rates for High and Low Qualification Groups')
plt.legend()

# Show the plot
plt.show()
```
